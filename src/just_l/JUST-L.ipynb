# パズルゲーム JUST-L を解くアルゴリズムの開発 （作：髙橋）

---
---

## **問題文**

$H$ 行 $W$ 列の盤面に`J`, `U`, `S`, `T`, `L`からなる $N$ 個の文字 $S_1, ..., S_N$ が書かれています．
ここで，文字 $S_i$ の座標は $R_i$ 行 $C_i$ 列です $(i = 1, ..., N)$．

あなたは，以下の形をした`J`, `U`, `S`, `T`, `L`のブロックを好きな数だけ選び，それらの向きを変えて盤面に置くことで，盤面に書かれた文字を全て隠したいです．
この際，ある文字の上に置かれるブロックはその文字の形をしたブロックである必要があります．

<img src="https://drive.google.com/uc?=view&id=1B4O2AKZPZFoxhU5JNFvMBqa9hk1XUKD-" width= 50%>

盤面が入力されるので，そのような置き方が存在するか判定して，存在する場合はそのような置き方の1つを求めてください．

---
---

## **制約**

- $H, W$ はそれぞれ3以上10以下の整数
- $N$ は2以上5以下の整数
- $S_i$ は`J`, `U`, `S`, `T`, `L`のいずれか $(i = 1, ..., N)$
- $R_i$ は1以上 $H$ 以下の整数 $(i = 1, ..., N)$
- $C_i$ は1以上 $W$ 以下の整数 $(i = 1, ..., N)$
- $R_i \neq R_j$ または $C_i \neq C_j$  $(1\leq i < j \leq N)$
- 各入力に対して，10秒以内にプログラムの実行が完了するようなアルゴリズムを記述すること

---
---

## **入出力**

### **入力**

入力は以下の形式でテキストファイルにより与えられる．
（examples/ に含まれる`in.txt`を参考にせよ．）

```
H W N
S_1 R_1 C_1
︙
S_N R_N C_N
```

### **出力**

- そのような置き方が存在しないとき，以下の形式で出力せよ．
```
No
```

- そのような置き方が存在するとき，盤面に置かれるブロックの数を $M$ ， $j = 1, ..., M$ 個目のブロックの形を $T_j$ ，その中心（各ブロックを含む 3 x 3 の正方形の中心）の座標を $X_j$ 行 $Y_j$ 列，その向き（上図参考；裏返しは不可）を $Z_j$ として以下の形式で出力せよ．ただし，同じマスに複数のブロックを置いてはいけない．また，盤面からはみ出すようにブロックを置いてはいけない．
```
Yes
M
T_1 X_1 Y_1 Z_1
︙
T_M X_M Y_M Z_M
```

テキストファイルに出力せよ．
（examples/ に含まれる`out.txt`を参考にせよ．TODO: この方法がよいのか先生と検討する．採点のしやすさに依存．）

## **発展要件**

題意を満たすブロックの置き方の中で，なるべく盤面を埋める置き方を考えてください．
具体的には，$K$ をブロックが置かれたマス目の数として，以下の式で定義される充填率を最大化するブロックの置き方を考えてください．

$$(充填率) := \frac{K}{HW} \times 100 ~[\%]$$

例えば，
- 出力例1-1の充填率は $28 / 48 \times 100 ≈ 58.3\%$ ，
- 出力例1-3の充填率は $33 / 48 \times 100 = 68.75\%$ ，
- 出力例3と出力例4の充填率は $100\%$

です．

発展要件の実行時間制限は**1分**とします．

**HINT**: この発展課題のように，ある値をなるべく良くする解を探索する手法のことをヒューリスティックと呼びます． 

上記の発展要件の他に，オリジナリティに富んだ発展課題も歓迎します．
